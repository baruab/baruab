```python

import networkx as nx
import matplotlib.pyplot as plt

g = nx.Graph()

g.add_node("Alpha")
g.add_edge("Alpha", "Bravo")
g.add_edge("Alpha", "Charlie")
g.add_edge("Alpha", "Foxtrot")
g.add_edge("Alpha", "Delta")

g.add_edge( "Bravo", "Echo")
g.add_edge( "Bravo", "Delta")
g.add_edge( "Bravo", "George")

g.add_edge( "Charlie", "Delta")
g.add_edge( "Charlie", "Foxtrot")

g.add_edge( "Delta", "Echo")
g.add_edge( "Delta", "George")
g.add_edge( "Delta", "Foxtrot")

g.add_edge( "Echo", "George")
g.add_edge( "Foxtrot", "Hawk")
g.add_edge( "Foxtrot", "George")

g.add_edge( "George", "Hawk")

g.add_edge("Hawk", "Indigo")
g.add_edge("Indigo", "Jack")


list(g.nodes())


pos = nx.spring_layout(g)

nx.draw_networkx(g, pos, node_color='#ffaaff', node_size=800)

plt.savefig("assignment1.png")
plt.show()

```
